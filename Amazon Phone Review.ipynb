# Task 2 Visualizations

### Rating Distribution

```python
sns.histplot(phone_review['rating_y'], bins = 10)
plt.title('Rating Distribution')
plt.savefig('Rating Distribution.png')
plt.show()
```

### Price vs Rating

```python
# Price vs Rating relationship
sns.scatterplot(data=phone_review, x='price', y = 'rating_y')
plt.title("Price vs Rating")
plt.savefig('Price vs Rating')
plt.show()
```

### Trend wise Rating and Total Reviews 

```python
plt.figure(figsize=(5,3))
phone_review['date']=pd.to_datetime(phone_review['date'])
phone_review.groupby(phone_review['date'].dt.month)['rating_y'].mean().plot()
plt.title("Trend wise Rating ")
plt.xlabel("Month")
plt.ylabel("Rating")
plt.savefig("Trend wise Rating.png")
plt.show()
plt.figure(figsize=(5,3))
phone_review.groupby(phone_review['date'].dt.month)['totalReviews'].sum().plot()
plt.title("Trend wise Total Reviews")
plt.xlabel("Month")
plt.ylabel("Total Reviews")
plt.savefig("Trend wise Total Reviews.png")
plt.show()
```

### Brand wise average rating , total reviews , average price 

```python
fig, axes = plt.subplots(3, 1, figsize=(8, 10))

# 1. Average Rating by Brand
brand_rating = phone_review.groupby('brand')['rating_y'].mean().reset_index()
brand_rating = brand_rating.sort_values('rating_y')
sns.barplot(data=brand_rating, x='rating_y', y='brand', ax=axes[0])
axes[0].set_title("Average Rating by Brand")

# 2. Total Reviews by Brand
brand_reviews = phone_review.groupby('brand')['totalReviews'].sum().reset_index()
brand_reviews = brand_reviews.sort_values('totalReviews')
sns.barplot(data=brand_reviews, x='totalReviews', y='brand', ax=axes[1])
axes[1].set_title("Total Reviews by Brand")

# 3. Average Price by Brand
brand_price = phone_review.groupby('brand')['price'].mean().reset_index()
brand_price = brand_price.sort_values('price')
sns.barplot(data=brand_price, x='price', y='brand', ax=axes[2])
axes[2].set_title("Average Price by Brand")

plt.tight_layout()
plt.savefig("Average Rating by Brand.png")
plt.show()
```

### Votes Rating wise

```python
phone_review['helpfulVotes'].describe()
phone_review.groupby('rating_y')['helpfulVotes'].mean().plot()
plt.xlabel("Average Rating")
plt.ylabel("helpfulVotes")
plt.title("Votes Rating wise")
plt.savefig("Votes Rating wise.png")
plt.show()
```

# Task 3. Sentiment Analysis

```python
pip install textblob
```

```python
from textblob import TextBlob
```

```python
def get_sentiment(text):
    if pd.isna(text):
        return 0
    return TextBlob(text).sentiment.polarity

phone_review['sentiment_score'] = phone_review['full_review'].apply(get_sentiment)
```

```python
def classify(score):
    if score > 0.1:
        return 'Positive'
    elif score < -0.1:
        return 'Negative'
    else:
        return 'Neutral'

phone_review['sentiment_label'] = phone_review['sentiment_score'].apply(classify)
```

```python
brand_sentiment = phone_review.groupby('brand')['sentiment_score'].mean().sort_values()
brand_sentiment.head()
```

```python
phone_review.groupby(['brand', 'sentiment_label']).size()
```

### Sentiment Distribution

```python
sns.countplot(data=phone_review, x='sentiment_label')
plt.title("Sentiment Distribution")
plt.savefig("Sentiment Distribution.png")
plt.show()
```

### Brand wise sentiment distribution

```python
sent_counts = phone_review.groupby(['brand', 'sentiment_label']).size().reset_index(name='count')

plt.figure(figsize=(12,6))
sns.barplot(data=sent_counts, x='brand', y='count', hue='sentiment_label')
plt.title("Brand-wise Sentiment Distribution")
plt.xticks(rotation=45)
plt.savefig("Brand wise Sentiment Distribution.png")
plt.show()
```
